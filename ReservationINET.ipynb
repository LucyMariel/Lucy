```python
# Sample data
x = df['target_datetime']

# Create the histogram
fig, ax = plt.subplots(figsize=(12, 8))
n, bins, patches = ax.hist(x, bins=21, edgecolor='white')

# Define custom labels
labels = ['2022-5', '2022-6', '2022-7', '2022-8', '2022-9', '2022-10',
          '2022-11', '2022-12', '2023-1', '2023-2', '2023-3', '2023-4',
          '2023-5', '2023-6', '2023-7', '2023-8', '2023-9', '2023-10',
          '2023-11', '2023-12', '2024-1']

# Ensure labels match the number of bins
if len(labels) != len(bins) - 1:
    raise ValueError("The number of labels must be one less than the number of bins.")

# Set x-ticks to match bin edges
ax.set_xticks(bins[:-1])  # Use only the edges that align with the labels

# Set x-tick labels
ax.set_xticklabels(labels, rotation=90, fontsize=10)
# Set axis labels and title
ax.set_xlabel('Date')  # Label for the x-axis
ax.set_ylabel('Frequency')  # Label for the y-axis
ax.set_title('Histogram of Reservation Number per Month')  # Title of the histogram
plt.show()
```
