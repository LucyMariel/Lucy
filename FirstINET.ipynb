```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("App_log.csv")
```

```python
df.loc[:, 'target_datetime'] = df.loc[:,'target_datetime'].apply(pd.to_datetime)
```

```python
print(df.dtypes)
```

```python
df['target_datetime'] = pd.to_datetime(df['target_datetime'])
```

```python
df.set_index('target_datetime', inplace=True)
```

```python
print(df.info())
```

```python
numeric_df = df.select_dtypes(include=[float, int])
```

```python
print(numeric_df.info())
```

```python
monthly_mean = numeric_df.groupby(pd.Grouper(freq='M')).mean()
```

```python
print(monthly_mean)
```

```python
monthly_count = numeric_df.groupby(pd.Grouper(freq='M')).count()
```

```python
print(monthly_count)
```
