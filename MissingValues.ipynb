```python
dates = pd.date_range(start="2021-01-01", end="2021-12-31")
dates
```

```python
df = pd.DataFrame(np.random.randint(1, 100, 365), index=dates, columns=["Random Value"])
df
```

A serie data was created with random values

```python
#for creating stadistical average
df.groupby(pd.Grouper(freq='M')).mean()
```

```python
df.resample('M').mean()
```

```python
dropna_df = df.dropna()
dropna_df
```

```python
df.mode()
df
```

```python
fill_freq_df = df.fillna(df.mode().iloc[0, :])
fill_freq_df
```
