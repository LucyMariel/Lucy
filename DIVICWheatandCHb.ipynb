**[Problem 2] Expansion to n × m mass**

Create a function that returns a ndarray containing the number of wheat on an nxm chess board. Then use this function to find the number of wheat on the 8x8 square chess board.

```python
%%timeit
import numpy as np

def wheat_on_chessboard(n, m):
    # Calculate total number of squares
    total_squares = n * m

    # Initialize the list with the first element
    wheat_list = [1]

    # Populate the list with the number of wheat grains on each square
    for _ in range(total_squares - 1):
        wheat_list.append(2 * wheat_list[-1])

    # Convert the list to a ndarray
    wheat_ndarray = np.array(wheat_list)

    # Reshape the ndarray to n x m
    wheat_ndarray = wheat_ndarray.reshape((n, m))

    return wheat_ndarray

# Use the function to find the number of wheat grains on an 8x8 chessboard
chessboard_8x8 = wheat_on_chessboard(8, 8)
print("Number of wheat grains on an 8x8 chessboard:")
print(chessboard_8x8)
```

**[Problem 3] Total number of whea**

Find the total number of wheat on a chess board of 8 x 8 squares. Also, find the average of each column and visualize it in a bar chart.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def wheat_on_chessboard(n, m):
    # Calculate total number of squares
    total_squares = n * m

    # Initialize the list with the first element
    wheat_list = [1]

    # Populate the list with the number of wheat grains on each square
    for _ in range(total_squares - 1):
        wheat_list.append(2 * wheat_list[-1])

    # Convert the list to a ndarray with integer type
    wheat_ndarray = np.array(wheat_list)

    # Reshape the ndarray to n x m
    wheat_ndarray = wheat_ndarray.reshape((n, m))

    return wheat_ndarray

# Use the function to find the number of wheat grains on an 8x8 chessboard
chessboard_8x8 = wheat_on_chessboard(8, 8)
print("Number of wheat grains on an 8x8 chessboard:")
print(chessboard_8x8)

# Calculate the average number of wheat grains in each column
column_averages = chessboard_8x8.mean(axis=0)

# Plot the averages of wheat grains in each column
plt.figure(figsize=(10, 6))
plt.bar(np.arange(1, 4), np.array([1, 2, 3]))
plt.xlabel("Column")
plt.ylabel("Number")
plt.title("Number in Each Column")
plt.show()
```

**[Problem 5] How many times is the second half as long as the first half?**

Find how many times the number of wheat in the second half of the chessboard is greater than the number of wheat in the first half. The first half is lines 0 to 3 of the array, and the second half is lines 4 to 7.

```python
import numpy as np

n_squares = 64
indices_of_squares = np.arange(n_squares).astype(np.uint64)
board_ndarray = 2**indices_of_squares

# Reshape the array to an 8x8 chessboard
board_ndarray = board_ndarray.reshape((8, 8))

# Calculate the sum of wheat grains in the first half (lines 0 to 3)
first_half_sum = board_ndarray[:4, :].sum()

# Calculate the sum of wheat grains in the second half (lines 4 to 7)
second_half_sum = board_ndarray[4:, :].sum()

# Calculate the ratio
ratio = second_half_sum / first_half_sum

print("Number of wheat in the first half of the chessboard:", first_half_sum)
print("Number of wheat in the second half of the chessboard:", second_half_sum)
print("The number of wheat in the second half is {} times greater than in the first half.".format(ratio))
```

**[Problem 6] Expansion to n × m mass by another calculation method**

Create a new function similar to Problem 2 that returns an ndarray containing the number of wheat squares on an n×m chessboard, one usingnp.append ()and one using broadcast.

Use those functions again to find the number of wheat on an 8x8 square chess board.

**np.append ()**

```python
%%timeit
import numpy as np

def wheat_on_chessboard(n, m):
    # Calculate total number of squares
    total_squares = n * m

    # Initialize the numpy array with the first element
    wheat_ndarray = np.array([1], dtype=np.int64)

    # Populate the numpy array with the number of wheat grains on each square
    for _ in range(total_squares - 1):
        wheat_ndarray = np.append(wheat_ndarray, 2 * wheat_ndarray[-1])

    # Reshape the ndarray to n x m
    wheat_ndarray = wheat_ndarray.reshape((n, m))

    return wheat_ndarray

# Use the function to find the number of wheat grains on an 8x8 chessboard
chessboard_8x8 = wheat_on_chessboard(8, 8)
print("Number of wheat grains on an 8x8 chessboard:")
print(chessboard_8x8)
```

**using broadcast**

```python
%%timeit
import numpy as np

def wheat_on_chessboard(n, m):
    # Calculate total number of squares
    total_squares = n * m

    # Create an array of indices from 0 to total_squares - 1
    indices = np.arange(total_squares)

    # Calculate the number of wheat grains on each square using broadcasting
    wheat_ndarray = 2**indices

    # Reshape the ndarray to n x m
    wheat_ndarray = wheat_ndarray.reshape((n, m))

    return wheat_ndarray

# Use the function to find the number of wheat grains on an 8x8 chessboard
chessboard_8x8 = wheat_on_chessboard(8, 8)
print("Number of wheat grains on an 8x8 chessboard:")
print(chessboard_8x8)

```
