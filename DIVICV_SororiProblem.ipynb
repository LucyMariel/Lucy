```python
import matplotlib.pyplot as plt

# Define the function to compute the rice grains
def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    -----------
    n_days: int
        Number of days to get rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice grains received each day (list length equals n_days)
    list_total_grains : list
        List of the total number of rice grains received by the end of each day (list length equals n_days)
    """

    list_n_grains = [2**i for i in range(n_days)]
    list_total_grains = [sum(list_n_grains[:i+1]) for i in range(n_days)]

    return list_n_grains, list_total_grains

# Call the function with the number of days the course lasts (e.g., 10 days)
n_days = 10
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=n_days)

# Print the results
print(f"Number of grains each day for {n_days} days:", list_n_grains)
print(f"Total number of grains by each day for {n_days} days:", list_total_grains)

# Plot the line graph
plt.figure(figsize=(10, 6))

# Plot grains received each day
plt.subplot(2, 1, 1)
plt.plot(range(1, n_days + 1), list_n_grains, marker='o', linestyle='-', color='b')
plt.xlabel('Day')
plt.ylabel('Rice Grains Received')
plt.title('Rice Grains Received Each Day')
plt.grid(True)

# Plot cumulative total of grains by each day
plt.subplot(2, 1, 2)
plt.plot(range(1, n_days + 1), list_total_grains, marker='o', linestyle='-', color='g')
plt.xlabel('Day')
plt.ylabel('Total Rice Grains')
plt.title('Cumulative Total of Rice Grains by Each Day')
plt.grid(True)

plt.tight_layout()
plt.show()
```
