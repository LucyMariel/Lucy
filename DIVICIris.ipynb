```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Step 2: Store the explanatory variables in a Pandas DataFrame
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)
Y = pd.DataFrame(data=iris.target, columns=['Species'])
df = pd.concat([X, Y], axis=1)

# Define a mapping dictionary
label_mapping = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}

# Apply the mapping to create new labels
df['Species'] = df['Species'].map(label_mapping)

# Calculate the number of samples per label after mapping
label_counts_mapped = df['Species'].value_counts()

# Plotting the pie chart with the updated labels
plt.figure(figsize=(8, 6))
label_counts_mapped.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Distribution of Iris Species')
plt.ylabel('')
plt.show()

print(df['Species'].value_counts())  # Print the count of each new label
```

Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Step 2: Store the explanatory variables in a Pandas DataFrame
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)
Y = pd.DataFrame(data=iris.target, columns=['Species'])
df = pd.concat([X, Y], axis=1)

# Define a mapping dictionary
label_mapping = {0: 'Label1', 1: 'Label2', 2: 'Label3'}

# Apply the mapping to create new labels
df['Species'] = df['Species'].map(label_mapping)

# Iterate over each feature and create box plots for each
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    plt.boxplot([df[df['Species'] == 'Label1'][feature],
                 df[df['Species'] == 'Label2'][feature],
                 df[df['Species'] == 'Label3'][feature]],
                labels=['Label1', 'Label2', 'Label3'])
    plt.title(f'Distribution of {feature} by Species')
    plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

Violin Plot Box

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Step 2: Store the explanatory variables in a Pandas DataFrame
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)
Y = pd.DataFrame(data=iris.target, columns=['Species'])
df = pd.concat([X, Y], axis=1)

# Define a mapping dictionary
label_mapping = {0: 'Label1', 1: 'Label2', 2: 'Label3'}

# Apply the mapping to create new labels
df['Species'] = df['Species'].map(label_mapping)

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Iterate over each feature and create violin plots for each
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='Species', y=feature, data=df, palette='muted')
    plt.title(f'Distribution of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

Box plots and violin plots are both used to visualize the distribution of data, but they differ in their approach and the insights they provide.

A box plot summarizes the distribution of a numerical variable through its quartiles.A violin plot combines aspects of a box plot and a kernel density plot.

Advantages Box Plot:

Simplicity: Easy to interpret and understand.
Outlier Detection: Clearly indicates outliers beyond the whiskers.
Comparative Analysis: Easy comparison between different groups or categories.

Advantages Violin Plot

Distribution Shape: Provides a more detailed view of the data distribution compared to box plots.
Density Estimation: Shows the density of the data at different values, giving a sense of data concentration.
Compact Representation: Combines multiple aspects (summary statistics and density) into a single plot.


Comparison:
Detail: Violin plots provide more information about the data distribution due to the density plot inside.
Interpretation: Box plots are easier to interpret at a glance, especially for comparing medians, quartiles, and outliers.
Visual Appeal: Violin plots can be more visually appealing and informative for complex or large datasets.
Usage Context: Box plots are more traditional and widely used for basic statistical comparisons, while violin plots are preferred when deeper insights into the distribution shape are needed.

**[Problem 7] Confirming the relationship between features**

Let's look at the relationship between features. Please do the following four things.

Make scatter plots of all combinations (six), such as choosing two from four features, color-coded by type.
The following figure is an example of one of them.
Create a scatterplot matrix that displays all combinations of features at once.

Create a scatterplot matrix that displays all combinations of features at once.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Step 2: Store the explanatory variables in a Pandas DataFrame
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)
Y = pd.DataFrame(data=iris.target, columns=['Species'])
df = pd.concat([X, Y], axis=1)

# Define a mapping dictionary
label_mapping = {0: 'Label1', 1: 'Label2', 2: 'Label3'}

# Apply the mapping to create new labels
df['Species'] = df['Species'].map(label_mapping)

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create scatter plots for each combination of features
feature_combinations = [('sepal length (cm)', 'sepal width (cm)'),
                        ('sepal length (cm)', 'petal length (cm)'),
                        ('sepal length (cm)', 'petal width (cm)'),
                        ('sepal width (cm)', 'petal length (cm)'),
                        ('sepal width (cm)', 'petal width (cm)'),
                        ('petal length (cm)', 'petal width (cm)')]

for i, (feat1, feat2) in enumerate(feature_combinations, 1):
    plt.subplot(3, 2, i)
    sns.scatterplot(x=feat1, y=feat2, hue='Species', data=df, palette='Set2')
    plt.title(f'Scatter plot: {feat1} vs {feat2}')
    plt.xlabel(feat1)
    plt.ylabel(feat2)
    plt.legend()

plt.tight_layout()
plt.show()

# Scatterplot matrix
sns.pairplot(df, hue='Species', palette='Set2')
plt.suptitle('Scatterplot Matrix of Iris Dataset Features', y=1.02)
plt.show()
```

Make a heat map of the correlation coefficient matrix (1)

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Step 2: Store the explanatory variables in a Pandas DataFrame
X = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Calculate the correlation matrix
corr_matrix = X.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Coefficient Matrix')
plt.show()
```
