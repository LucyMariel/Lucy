. Multidimensional normally distributed random numbers can be created by using np.random.multivariate_normal() .
numpy.random.multivariate_normal â€” NumPy v1.19 Manual

The shape of the created ndarray will be (500, 2).

<Seed value>

To ensure that the value does not change from one run to the next, we will fix the seed value of the random number using np.random.seed() . We will use a seed value of 0.

```python
import numpy as np
# Set the seed value for reproducibility
np.random.seed(0)

mean = (-3, 0)
cov = [[1.0, 0.8], [0.8, 1.0]]

# Generate 500 random numbers with a two-dimensional normal distribution
x = np.random.multivariate_normal(mean, cov, 500)

# Print the shape to confirm it's (500, 2)
print(x.shape)
```

**[Problem 6] Labeling**

For each of the 1000 pieces of data that you have combined together, label them to indicate whether they are from question 1 or 4. Question 1 should be labeled 0 and Question 4 should be labeled 1 .

Create a ndarray of (1000, 3) by adding a new column of labels containing 0s and 1s.

Datasets used for machine learning often have this form.

<< Creation example >>

When there are 3 + 3 data, the ndarray is (6, 3) as below.

```python
import numpy as np

# Set the seed value for reproducibility for the first set
np.random.seed(0)

# Define the first mean and covariance matrix
mean1 = [-3, 0]
cov1 = [[1.0, 0.8], [0.8, 1.0]]

# Generate the first set of 500 random numbers with a two-dimensional normal distribution
x = np.random.multivariate_normal(mean1, cov1, 500)

# Set the seed value for reproducibility for the second set
np.random.seed(0)

# Define the second mean and covariance matrix
mean2 = [0, -3]
cov2 = [[1.0, 0.8], [0.8, 1.0]]

# Generate the second set of 500 random numbers with a two-dimensional normal distribution
y = np.random.multivariate_normal(mean2, cov2, 500)

# Concatenate the two datasets
df = np.concatenate([x, y], axis=0)

# Create the labels
labels = np.concatenate([np.zeros(500), np.ones(500)])

# Combine the data and labels into a single ndarray
df_with_labels = np.column_stack((df, labels))

# Print the shape to confirm it's (1000, 3)
print(df_with_labels.shape)

# Display the first few rows to check
print(df_with_labels[:10])
```
