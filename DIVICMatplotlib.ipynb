```python
x1 = [1, 2, 3]
y1 = [2, 4, 6]
x2 = [1, 2, 3]
y2 = [-2, -4, -6]

fig, axes = plt.subplots(1, 2)
axes[0].plot(x1, y1)
axes[1].plot(x2, y2)
axes[0].set_title('Subplot1 title')
axes[1].set_title('Subplot2 title')
fig.suptitle('Graph title')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
matplotlib.style.available
```

```python
df = pd.read_csv('App_log.csv')
```

```python
df.columns
```

```python
df.dtypes
```

```python
x = df['reservation_no']
y = df['estimated_amount']
plt.plot(x, y)
plt.show()
```

```python
x = df['reservation_no']
y = df['estimated_amount']
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x = df['reservation_no']
y = df['estimated_amount']
matplotlib.style.use('classic')
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

```python
x = df['reservation_no']
y = df['estimated_amount']
matplotlib.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots()
ax.plot(x, y)
fig.suptitle('Relation Reservation-Cost')
ax.set_xlabel('reservation number')
ax.set_ylabel('estimated amount')
ax.plot(x, y, label='sample legend')
ax.legend(loc='center')
plt.show()
```

```python
x = df['reservation_no']
y = df['estimated_amount']
matplotlib.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots()
ax.bar(x, y)
fig.suptitle('Relation Reservation-Cost')
ax.set_xlabel('reservation number')
ax.set_ylabel('estimated amount')
ax.plot(x, y, label='sample legend')
ax.legend(loc='center')
plt.show()
```
