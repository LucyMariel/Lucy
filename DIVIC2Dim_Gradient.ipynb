**[Problem 3] Find the gradient**

Find the gradient at each x-coordinate. Here, the gradient is calculated by dividing the change in y by the change in x.

Gradient = Amount of change in Y/ Amount of change in X

Each amount of change is approximately calculated from the difference between adjacent elements as shown in the figure below.

```python
import numpy as np

# Create the x and y arrays
x = np.arange(-50, 50, 0.1)
y = 1/2 * x + 1

# Combine x and y into a single array of shape (1001, 2)
combined_array = np.column_stack((x, y))

# Compute the gradient of y with respect to x
gradient = np.gradient(y, x)

print("Combined array:")
print(combined_array)
print("\nGradient of y with respect to x:")
print(gradient)
```

**[Problem 6] Find the minimum value**

Find the minimum value of y for the function treated in Problem 5. The range of x is the same as in Problem 5.

This time, use the ndarray min method and argmin method instead of the gradient descent method. The min method can get the minimum value, and the argmin method can get the index of the minimum value.

Also, display the slope before and after taking the minimum y.

link: numpy.ndarray.min — NumPy v1.16 Manual

link: numpy.ndarray.argmin — NumPy v1.16 Manual

```python
import numpy as np
import matplotlib.pyplot as plt

# Function definitions
def function1(x):
    return x**2

def function2(x):
    return 2*x**2 + 2*x

def function3(x):
    return np.sin(np.sqrt(x))

# Define ranges for x
x1_range = np.arange(-50, 50, 0.1)
x2_range = np.arange(-50, 50, 0.1)
x3_range = np.arange(0.1, 50, 0.1)  # Starting from 0.1 to avoid sqrt(0)

# Calculate y values for each function
y1 = function1(x1_range)
y2 = function2(x2_range)
y3 = function3(x3_range)

# Find minimum y values and their corresponding x values
min_y1 = np.min(y1)
min_y2 = np.min(y2)
min_y3 = np.min(y3)

min_x1 = x1_range[np.argmin(y1)]
min_x2 = x2_range[np.argmin(y2)]
min_x3 = x3_range[np.argmin(y3)]

# Find the slopes before and after taking the minimum y
slope_before_min_y1 = (function1(min_x1 + 0.1) - function1(min_x1)) / 0.1
slope_after_min_y1 = (function1(min_x1) - function1(min_x1 - 0.1)) / 0.1

slope_before_min_y2 = (function2(min_x2 + 0.1) - function2(min_x2)) / 0.1
slope_after_min_y2 = (function2(min_x2) - function2(min_x2 - 0.1)) / 0.1

slope_before_min_y3 = (function3(min_x3 + 0.1) - function3(min_x3)) / 0.1
slope_after_min_y3 = (function3(min_x3) - function3(min_x3 - 0.1)) / 0.1

# Print results
print("Minimum y values:")
print("Function 1: y =", min_y1, "at x =", min_x1)
print("Function 2: y =", min_y2, "at x =", min_x2)
print("Function 3: y =", min_y3, "at x =", min_x3)

print("\nSlope before and after minimum y:")
print("Function 1: Slope before =", slope_before_min_y1, ", Slope after =", slope_after_min_y1)
print("Function 2: Slope before =", slope_before_min_y2, ", Slope after =", slope_after_min_y2)
print("Function 3: Slope before =", slope_before_min_y3, ", Slope after =", slope_after_min_y3)

# Plotting the functions
plt.figure(figsize=(14, 10))

# Function 1
plt.subplot(3, 1, 1)
plt.plot(x1_range, y1, label='y = x^2')
plt.scatter(min_x1, min_y1, color='red', label='Minimum y')
plt.title('Function 1')
plt.legend()

# Function 2
plt.subplot(3, 1, 2)
plt.plot(x2_range, y2, label='y = 2*x^2 + 2*x')
plt.scatter(min_x2, min_y2, color='red', label='Minimum y')
plt.title('Function 2')
plt.legend()

# Function 3
plt.subplot(3, 1, 3)
plt.plot(x3_range, y3, label='y = sin(sqrt(x))')
plt.scatter(min_x3, min_y3, color='red', label='Minimum y')
plt.title('Function 3')
plt.legend()

plt.tight_layout()
plt.show()
```
